```python
# ============================================
#  Ejercicio 9.1 - Agenda de contactos en Gradio (para Colab)
#  Celda única - versión corregida
# ============================================

# Si Gradio no está instalado, descomenta esta línea y ejecuta:
# !pip install gradio

import gradio as gr

def agregar_contacto(nombres, apellidos, fecha, direccion, telefono, correo, estado):
    """
    estado: lista de filas [[nombres, apellidos, fecha, direccion, telefono, correo], ...]
    Devuelve: nuevo_estado, tabla_para_mostrar, mensaje
    """

    # Asegurarnos de que haya una lista
    if estado is None:
        estado = []

    # Normalizar textos
    nombres   = (nombres or "").strip()
    apellidos = (apellidos or "").strip()
    fecha     = (fecha or "").strip()
    direccion = (direccion or "").strip()
    telefono  = (telefono or "").strip()
    correo    = (correo or "").strip()

    # Validación básica
    if not nombres or not apellidos or not fecha or not direccion or not telefono or not correo:
        return estado, estado, "⚠️ Todos los campos son obligatorios (incluida la fecha)."

    # Crear nueva fila de contacto
    nueva_fila = [nombres, apellidos, fecha, direccion, telefono, correo]

    # Nuevo estado (clonando la lista)
    nuevo_estado = estado + [nueva_fila]

    return nuevo_estado, nuevo_estado, "✅ Contacto agregado correctamente."


with gr.Blocks(title="Agenda de contactos - Ejercicio 9.1") as demo:
    gr.Markdown("## Agenda de contactos (Ejercicio 9.1)\nIntroduce los datos y pulsa **Agregar contacto**.")

    with gr.Row():
        with gr.Column():
            nombres_in = gr.Textbox(label="Nombres")
            apellidos_in = gr.Textbox(label="Apellidos")
            fecha_in = gr.Textbox(label="Fecha de nacimiento (por ejemplo: 13/05/2004)")
            direccion_in = gr.Textbox(label="Dirección")
            telefono_in = gr.Textbox(label="Teléfono")
            correo_in = gr.Textbox(label="Correo electrónico")
            btn_agregar = gr.Button("Agregar contacto")

        with gr.Column():
            tabla = gr.Dataframe(
                headers=["Nombres", "Apellidos", "Fecha nac.", "Dirección", "Teléfono", "Correo"],
                datatype=["str", "str", "str", "str", "str", "str"],
                value=[],          # empezamos con tabla vacía
                interactive=False,
                label="Contactos registrados"
            )

    mensaje = gr.Markdown("")  # Para mostrar errores o confirmaciones

    # Estado interno donde guardamos la lista de contactos
    estado = gr.State([])

    # Conectar botón con la función
    btn_agregar.click(
        fn=agregar_contacto,
        inputs=[nombres_in, apellidos_in, fecha_in, direccion_in, telefono_in, correo_in, estado],
        outputs=[estado, tabla, mensaje]
    )

demo.launch()
```
